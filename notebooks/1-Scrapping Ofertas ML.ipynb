```python
d={}
name = str(art1.find_all('h1', class_='ui-pdp-title')[0])
name = name[name.find('"title">')+26:-5]
price = str(art1.find_all('span', class_='andes-money-amount__fraction')[0])
price = float(price[price.find('">"')+63:-7].replace(',',''))
new_price = str(art1.find_all('meta', itemprop='price')[0])
new_price = float(new_price[new_price.find('"content"')+16:-20].replace(',',''))
```

```python
d={}
l=[]
for x in range(10):
    articulo = BeautifulSoup(requests.get(links[x]).content,'html.parser')
    name = str(articulo.find_all('h1', class_='ui-pdp-title')[0])
    price = str(articulo.find_all('span', class_='andes-money-amount__fraction')[0])
    new_price = str(articulo.find_all('meta', itemprop='price')[0])
    d['nombre'] = name[name.find('"title">')+26:-5]
    d['precio'] = float(price[price.find('">"')+63:-7].replace(',',''))
    d['precio nuevo'] = float(new_price[new_price.find('"content"')+16:-20].replace(',',''))
    pd.DataFrame(d,index=[0])
    l.append(pd.DataFrame(d,index=[x]))
```

```python
mas_vendidos=pd.concat(l)
```
